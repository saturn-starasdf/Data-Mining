## TRANSFORMATION

```python
# 04. Data transformation ##########################################################
    
loans_transform=loans.copy(deep=True)   
# print out and display dataframe as tables in HTML
display(HTML(loans_transform.head(5).to_html()))

# convert categorical data to numerical data, e.g., Degree #########################

print(loans_transform['Years in current job'].dtype)
print(loans_transform['Home Ownership'].dtype)
print(loans_transform['Purpose'].dtype)

df_dummies_years=pd.get_dummies(loans_transform['Years in current job'])
df_dummies_ownership=pd.get_dummies(loans_transform['Home Ownership'])
df_dummies_purpose=pd.get_dummies(loans_transform['Purpose'])

# add binary variables to dataframe
loans_transform=loans_transform.join(df_dummies_years)
loans_transform=loans_transform.join(df_dummies_ownership)
loans_transform=loans_transform.join(df_dummies_purpose)

# remove the original categorical variable
loans_transform=loans_transform.drop('Years in current job',1)
loans_transform=loans_transform.drop('Home Ownership',1)
loans_transform=loans_transform.drop('Purpose',1)

loans_transform=loans_transform.replace('Fully Paid',1)
loans_transform=loans_transform.replace('Charged Off',0)

loans_transform=loans_transform.drop('1 year',1)
loans_transform=loans_transform.drop('HaveMortgage',1)
loans_transform=loans_transform.drop('Home Improvements',1)

display(HTML(loans_transform.head(5).to_html()))
```

## NORMALIZATION

```python
# 03. Normalization ################################################################################

# find numeric columns
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
# get column names
cols_numeric = loans_transform.select_dtypes(include=numerics).columns.tolist()
# get column indices
cols_numeric_index=[loans_transform.columns.get_loc(col) for col in cols_numeric]
print('Numerical column names:\n',cols_numeric)
print('Numerical column indices:\n',cols_numeric_index)
    
# create a copy first
df_norm=loans_transform.copy(deep=True)

# Normalization method 1 = min-max normalization
# API, https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
loans_transform[cols_numeric]=scaler.fit_transform(loans_transform[cols_numeric])
display(HTML(loans_transform.head(10).to_html()))

# Normalization method 2 = Min-max normalization [0,1]
for col in cols_numeric:
    df_norm[col]=(loans_transform[col]-loans_transform[col].min())/(loans_transform[col].max()-loans_transform[col].min())
```

## KNN

```python
#KNN (Distance Measure: Euclidian distance)
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

y = df_norm['Term'] # define label as nominal values
le = preprocessing.LabelEncoder()
le.fit(y)
y_encoded = le.transform(y)

df_norm['Term'] = y_encoded
x = df_norm.drop('Term',1)
y = df_norm['Term'] 

display(HTML(df_norm.head(10).to_html()))

x_train, x_test, y_train, y_test = train_test_split(x, y_encoded, 
                                                    test_size=0.2)

# build and eval models
from sklearn import neighbors
from sklearn.metrics import accuracy_score
# API， https://scikit-learn.org/stable/modules/generated/sklearn.metrics.precision_score.html
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from matplotlib import pyplot as plt

# API for KNeighborsClassifier
# https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html

for k in range(5, 10, 2): 
    clf=neighbors.KNeighborsClassifier(k, weights='uniform')
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    print('K =', k, ', Accuracy: ', round(accuracy_score(y_test, y_pred),3), ', Precision: ', round(precision_score(y_test, y_pred, average='micro'),3),', Recall: ', round(recall_score(y_test, y_pred, average='micro'),3), ', F1 Score: ', round(f1_score(y_test,y_pred),3))

# By N-fold cross evaluations ###############################################
from sklearn.model_selection import cross_val_score

for k in range(5, 10, 2): 
    clf=neighbors.KNeighborsClassifier(k, weights='uniform')
    acc=cross_val_score(clf, x, y, cv=5, scoring='accuracy').mean()
    pcs=cross_val_score(clf, x, y, cv=5, scoring='precision').mean()
    rcl=cross_val_score(clf, x, y, cv=5, scoring='recall').mean()
    f1=cross_val_score(clf, x, y, cv=5, scoring='f1').mean()
    print('K =', k, ', Accuracy: ',round(acc,3),', Precision', round(pcs,3),', recall', round(rcl,3),', f1',round(f1,3))
```
